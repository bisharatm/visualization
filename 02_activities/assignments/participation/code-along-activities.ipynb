### 1 - Making simple figures with matplotlib

```python
# Import the necessary libraries.

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Generate sample data with a fixed random seed for reproducibility.

np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
# Produce a scatter plot of the sample data.

fix, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y)
```

```python
# Produce a scatter plot of the sample data.

fix, ax = plt.subplots(figsize=(5, 3))
ax.bar(x, y)
```

```python
# Produce a scatter plot of the sample data.

fix, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
```

```python
# Activity: Modify above code to produce a historgram.

fix, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

### 2 - Plot title and labels

```python
# Add axes title and labels to the plot.
fix, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
# Customize fonts for title and labels of the plot.

font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':15}

fix, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)

ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
```

```python
# Modify title and label positions on the plot.
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title('Total growth over time', fontdict=font1, loc='right')
ax.set_ylabel('Total growth', fontdict=font2, loc='bottom')
ax.set_xlabel('Years since start', fontdict=font2, loc='left')
```

### 3 - Customizing plot appearance

```python
# Change marker style and color in scatter plot.

fix, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y, marker='*', color='indigo')
```

```python
# Modify line style, width, and color in line plot.
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker='*', color='indigo', linestyle='--', linewidth=2)
```

```python
# Further customize line and marker appearance in line plot.

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y,
        marker='*',
        markersize=12,
        color='indigo',
        linestyle='--',
        linewidth=2,
        markeredgecolor='#fa9359',
        markerfacecolor='#000000')


# Add grid lines to the plot.
ax.grid(axis='y', color='blue', linestyle='--', linewidth=0.5)
```

#### Activity: Exploring matplotlib

```python
# Activity notes:
# The plot below uses the seaborn library & dataset to visualize the
# correlation between pairs of numeric features in the iris dataset,
# differentiated by species label. The visualization has good aesthetic,
# substantive, and perpectual qualities.

# Import libraries and load dataset.
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')

# Create pair plot with customized markers and color palette.
sns.pairplot(df, kind="scatter", hue="species", markers=["o", "s", "D"], palette="Set2")
plt.show()
```
