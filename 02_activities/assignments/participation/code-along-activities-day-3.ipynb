```python
# Import the necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Generate sample data

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
# Activity: Create a line plot with the two y variables against x

fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x, y1, color='blue', linestyle='-')
ax.plot(x, y2, color='orange', linestyle='-')
```

#### Legends

```python
# Add a legend to differentiate the two lines
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label='Person 1', color='blue', linestyle='-')
ax.plot(x, y2, label='Person 2', color='orange', linestyle='-')
ax.legend(loc='lower right')
```

```python
# Customize the legend's appearance
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label='Person 1', color='blue', linestyle='-')
ax.plot(x, y2, label='Person 2', color='orange', linestyle='-')
ax.legend(loc='lower right',
          frameon=True,     # add a frame around the legend
          fontsize=12,      # set font size
          ncol=2,           # set number of cols to fit legend
          shadow=True)
```

```python
# Move the legend outside of the plot area

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1, label='Person 1', color='blue', linestyle='-')
ax.plot(x, y2, label='Person 2', color='orange', linestyle='-')

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

#### Subplots

```python
# Generate sample data for subplots
np.random.seed(613)

x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)

x2 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# Create two subplots, one for each of the above datasets
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(7, 3))
ax1.scatter(x1, y1)
ax2.bar(x2, y2)
```

```python
# Activity: Customize the subplots further
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))

ax1.scatter(x1, y1, color='green')
ax1.set_title('Plot Title')
ax1.set_xlabel('X Axis Label')
ax1.set_ylabel('Y Axis Label')

ax2.bar(x2, y2, color='orange')
ax2.set_title('Chart Title')
ax2.set_xlabel('Character')
ax2.set_ylabel('Value')
```

```python
# Use subplot_mosaic() to create more complex subplot layouts
fig, someaxes = plt.subplot_mosaic(
    [['ax1', 'ax3'],     # ax3 spans two rows
     ['ax2', 'ax3']],
    figsize=(7, 4))

someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1, y1)
```

#### Modifying figure layout

```python
# Add some large axis titles to our previous plot
# Resulting plot has axis labels not fitting well within the figure area

fig, someaxes = plt.subplot_mosaic([
        ['ax1', 'ax3'],
        ['ax2', 'ax3']
    ], figsize=(7, 4)
)

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18)
```

```python
# Use layout parameter to adjust spacing automatically
fig, someaxes = plt.subplot_mosaic([
        ['ax1', 'ax3'],
        ['ax2', 'ax3']
    ], figsize=(7, 4),
    layout='constrained'    # automatically adjust spacing to ensure everything fits
)

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18)
```

#### Multiple visualisations on one axes

```python
# Create sample data for this activity
x = np.array(['Luffy', 'Zoro', 'Nami', 'Usoop', 'Sanjo'])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# Define figure and axes
fig, ax = plt.subplots(figsize=(7,3))

# Call to plot y1 and y2 agains x on the same axes
ax.bar(x, y1,
color = "indigo")
ax.plot(x, y2,
color = "red")
```

```python
# Add error information to a plot

# Calculate standard deviation for y2
y2_sd = np.std(y2)

# Create the plot of y2 vs x
fig, ax = plt.subplots(figsize=(7,3))
ax.bar(x, y2, color = "red")

# Add error bars to the bar plot
ax.errorbar(x, y2, yerr=y2_sd, fmt='none')
```

```python
# Customize error bar appearance
fig, ax = plt.subplots(figsize=(7, 3))

ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2,
        yerr = y2_sd,
        fmt = 'none',
        ecolor= 'indigo',
        elinewidth=4,
        capsize=6,
        capthick=4,
        errorevery=2) # show error bar for every 2nd data point
```

#### Add images to plots

```python
# Import necessary libraries
from PIL import Image   # to open images
import requests         # to get images from URLs
from io import BytesIO  # to store images
```

```python
# Get the image from a URL
response = requests.get(
    'https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png',

    # add to fix error due to response status code and content type
    headers={'User-Agent': 'Mozilla/5.0'}

)

image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
# Recreate the plot
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color='red')

# Add (overlay) a new axes for the image on the figure
ax_image = fig.add_axes(
    [0.1,   # x coorinate of bottom left corner (on figure, not axes)
     0.11,  # y coordinate of bottom left corner (on figure, not axes)
     0.15,  # image width (fraction of figure width)
     0.25]  # image height (fraction of figure height
)

# Add image to the new axes, and remove x/y axis
ax_image.imshow(image)
ax_image.axis('off')
```
